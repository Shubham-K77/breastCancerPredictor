<h2 style="font-weight: 600; text-align: center">4. Data Preprocessing</h2>

```python
data['diagnosis'].value_counts()
```

**The target output is categorical, so this is a classification problem. Machine learning models can’t understand labels like Benign or Malignant, so we need to convert them into numerical values. For two-class outputs, we use binary encoding. For features where the order matters, like First or Second, we use ordinal encoding. And for features with multiple unordered categories, one-hot encoding is the preferred method. These encoding techniques make the data understandable and usable for ML models.**

```python
data['diagnosis'] = data['diagnosis'].apply(lambda x: 1 if x == 'M' else 0)
```

**After binary encoding, Malignant cells are represented as 1 and Benign cells as 0. This aligns with the goal of cancer detection: 1 indicates the presence of a cancerous cell, while 0 indicates a non-cancerous cell.**

```python
data['diagnosis'].value_counts()
```

**The data after the binary encoding.**

```python
data.info()
```

**Here, now finally every feature has a numerical value meaning that they are ready for further processing.**

```python
data.hist(figsize=(18, 16))
```

**The histogram illustrates the distribution of feature values.**

```python
data.skew()
```

**The histogram shows that most features are right-skewed, meaning most data points are clustered on the left with a long tail to the right. For example, in income data, most individuals earn below the average, while few earn much higher. Machine learning models perform best when data is roughly normally distributed (skewness between -0.5 and 0.5).**

```python
from sklearn.preprocessing import PowerTransformer
pt = PowerTransformer(method='yeo-johnson')
data[["radius_mean", "texture_mean", "perimeter_mean", "area_mean", "compactness_mean", "concavity_mean", "concave points_mean", "symmetry_mean", "fractal_dimension_mean", "radius_se", "texture_se", "perimeter_se", "area_se", "smoothness_se", "compactness_se", "concavity_se", "concave points_se", "symmetry_se", "fractal_dimension_se", "radius_worst", "perimeter_worst", "area_worst", "compactness_worst", "concavity_worst", "symmetry_worst", "fractal_dimension_worst"]] = pt.fit_transform(data[["radius_mean", "texture_mean", "perimeter_mean", "area_mean", "compactness_mean", "concavity_mean", "concave points_mean", "symmetry_mean", "fractal_dimension_mean", "radius_se", "texture_se", "perimeter_se", "area_se", "smoothness_se", "compactness_se", "concavity_se", "concave points_se", "symmetry_se", "fractal_dimension_se", "radius_worst", "perimeter_worst", "area_worst", "compactness_worst", "concavity_worst", "symmetry_worst", "fractal_dimension_worst"]])
```

**The Yeo-Johnson transformation is a powerful technique for handling skewness. While methods like log transform address right (positive) skew and Box-Cox handles left (negative) skew, Yeo-Johnson can effectively correct both types of skewness, making the data more suitable for machine learning models.**

```python
data.skew()
```

**Here, every skewness value is between the -0.5 and 0.5 range making ot perfectly symmetric and suitable for the operations.**

```python
data.hist(figsize=(18, 16))
```

**Here, histogram proves the successful treatment of the skewness by Yeo-Johnson Method.**

<h2 style="font-weight: 600; text-align: center">6. Feature Engineering</h2>

```python
data['concavity_per_fractal_dimension_mean'] = data['concavity_mean'] / data['fractal_dimension_mean']
data['concavity_per_fractal_dimension_se'] = data['concavity_se'] / data['fractal_dimension_se']
data['concavity_per_fractal_dimension_worst'] = data['concavity_worst'] / data['fractal_dimension_worst']
```

**concavity_per_fractal_dimension**
<br />
**What it measures:** Measures indentations of the cell boundary relative to edge jaggedness.
<br />
**Relation to output:** Higher values may indicate irregular, highly indented, and jagged cells → higher likelihood of malignancy.


```python
data['compactness_per_texture_mean'] = data['compactness_mean'] / data['texture_mean']
data['compactness_per_texture_se'] = data['compactness_se'] / data['texture_se']
data['compactness_per_texture_worst'] = data['compactness_worst'] / data['texture_worst']
```

**compactness_per_texture**
<br />
**What it measures:** Density of cell particles relative to surface roughness.
<br />
**Relation to output:** Dense and rough cells often correspond to abnormal/malignant cells → may help detect cancer.


```python
data['concave_points_per_perimeter_mean'] = data['concave points_mean'] / data['perimeter_mean']
data['concave_points_per_perimeter_se'] = data['concave points_se'] / data['perimeter_se']
data['concave_points_per_perimeter_worst'] = data['concave points_worst'] / data['perimeter_worst']
```

**concave_points_per_perimeter**
<br />
**What it measures:** Number of inward dents normalized by total perimeter.
<br />
**Relation to output:** Cells with more concave points relative to their size are more likely to be malignant.


```python
data['compactness_per_area_mean'] = data['compactness_mean'] / data['area_mean']
data['compactness_per_area_se'] = data['compactness_se'] / data['area_se']
data['compactness_per_area_worst'] = data['compactness_worst'] / data['area_worst']
```

**compactness_per_area**
<br />
**What it measures:** Density of cell particles normalized by cell area.
<br />
**Relation to output:** Provides size-independent measure of compactness; higher values may indicate tightly packed, cancerous cells.

```python
data.skew()
```

**Here, the new features still are skewed and must be treated properly.**

```python
data[["concavity_per_fractal_dimension_mean", "concavity_per_fractal_dimension_se", "concavity_per_fractal_dimension_worst", "compactness_per_texture_mean", "compactness_per_texture_se", "compactness_per_texture_worst", "concave_points_per_perimeter_mean", "concave_points_per_perimeter_se", "concave_points_per_perimeter_worst", "compactness_per_area_mean", "compactness_per_area_se", "compactness_per_area_worst"]] = pt.fit_transform(data[["concavity_per_fractal_dimension_mean", "concavity_per_fractal_dimension_se", "concavity_per_fractal_dimension_worst", "compactness_per_texture_mean", "compactness_per_texture_se", "compactness_per_texture_worst", "concave_points_per_perimeter_mean", "concave_points_per_perimeter_se", "concave_points_per_perimeter_worst", "compactness_per_area_mean", "compactness_per_area_se", "compactness_per_area_worst"]])
```

<h2 style="font-weight: 600; text-align: center">7. Splitting and Scaling the data</h2>

```python
from sklearn.model_selection import train_test_split
x = data.drop(['diagnosis'], axis=1)
y = data['diagnosis']
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state=42)
```

**Splitting the data ensures that 80% is used for training the model, while the remaining 20% is reserved for evaluating its performance. Keeping a separate test set prevents overfitting and gives a realistic measure of accuracy. The random_state ensures the split is reproducible, so others get the same data separation.**

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

**Scaling is an important preprocessing step, but it is applied after feature engineering and skewness correction. The output (target) is not scaled. Scaling ensures that features measured on different scales, such as kilometers versus centimeters, are treated equally by the model. Without scaling, features with larger values could dominate the learning process, even if they are not more important.**
<br />

<ul>
    <li>fit_transform() calculates mean and std from the data it’s applied to.</li>
    <li>Using fit_transform() on X_test would leak information from the test set and is not correct. Therefore, we use transform as the calculation done previously will be properly utilized instead of creating a new one.</li>
</ul>

```python
from sklearn.linear_model import LogisticRegression
reg = LogisticRegression()
reg.fit(X_train, y_train)
```

**Logistic Regression is used to train the model and perform prediction tasks on the training data.**

**Now, the trained model has used the testing data to make predictions for the target values.**

```python
y_predict_logistic = pd.DataFrame({"Actual": y_test, "Predicted": y_predict})
y_predict_logistic.head(20)
```

**The first 20 predictions made by the model are all correct, matching the actual outcomes.**

Metrics are rounded to four decimal places to make comparison easier.

### Model Comparison and Conclusion

After benchmarking a broad mix of linear, distance-based, and tree ensembles for breast cancer diagnosis, a few clear themes stand out:

1. **Logistic Regression remains the most dependable choice.** It posts the top accuracy (0.9825) while keeping precision and recall perfectly aligned at 0.9767. That balance means clinicians receive confident alerts without noticeably increasing missed positives.

2. **KNN benefits greatly from the distance metric.** The Euclidean and Cosine variants both reach 0.9737 accuracy, but they lean into different strengths; Euclidean favours precision (0.9762) while Cosine edges ahead on recall (0.9767). The Manhattan flavour trades some precision (0.9111) for higher sensitivity, which could still be valuable when the cost of missing a malignant case outweighs the cost of a false alarm.

3. **Linear SVM mirrors the strong KNN results.** With 0.9737 accuracy and a precision–recall profile (≈0.9761/0.9535) similar to Euclidean KNN, it offers another high-performing, well-understood option that is easy to regularise and explain.

4. **Ensemble methods provide solid but not superior gains.** Random Forest and its tuned variant stabilise around 0.9649 accuracy, and XGBoost models hover between 0.9561 and 0.9649 depending on tuning. They are reliable workhorses, but on this dataset, they do not surpass the simpler baselines.

5. **Probabilistic simplicity still holds value.** Naive Bayes hits 0.9474 accuracy with perfectly matched precision, recall, and F1 (all 0.9302). It is lightweight, fast to retrain, and offers a transparent fallback when computing power or turnaround time is constrained.

6. **Key takeaways for deployment.** Every model achieves 0.94 accuracy, so the differentiator is how each balances false positives versus false negatives. Starting with Logistic Regression gives an interpretable, high-performing baseline, while Euclidean or Cosine KNN can act as safety nets in ensemble or triage settings. If operational constraints demand even faster inference, Naive Bayes remains serviceable, but be prepared to monitor precision.
